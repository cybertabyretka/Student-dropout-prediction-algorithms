```python
def get_num_cat_columns(df):
    numeric_columns = df.select_dtypes(include=[np.int64, np.float64]).columns.tolist()
    categorical_columns = df.select_dtypes(include=['object', 'category']).columns.tolist()
    return numeric_columns, categorical_columns
```

```python
num_cols_2024, cat_cols_2024 = get_num_cat_columns(data_2024_csv)
num_cols_2023, cat_cols_2023 = get_num_cat_columns(data_2023_csv)
```
